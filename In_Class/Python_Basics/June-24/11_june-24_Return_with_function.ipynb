### Return

- With return keyword we can get the same as global keyword 
- instead of print we just need to user return and it will work in the same way but just that when calling the function back we need give the same variable name as the required
- eg: 
    - def add ( ):
        - n1 = 10 
        - n2 = 12
        - sume = n1 + n2
        - return (sume,n1,n2)
        
    - sume,n1,n2 = add ( )
- with above equation we can understand the return and catch the variables 

```python
import time 
import math 
import random
```

```python
def add ():
    n1 = 10 
    n2 = 12
    sume = n1 + n2
    return (sume,n1,n2)
        
sume,n1,n2 = add ()
print (sume,n1,n2)
```
