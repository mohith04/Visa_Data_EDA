# If we need to import the any packages we will need to follow the steps
- Step 1 - import <package : name>
       - this will download the package 
       - eg : import random
            - import math 
            - import time
- Step 2 - dir (<package>)
       - this will get the contents or methods under the package 
       - eg : dir (random)
- Step 3 - help (<package>.<method>)
       - this will get the details of the method in the package
       - eg : help (random.randint)
