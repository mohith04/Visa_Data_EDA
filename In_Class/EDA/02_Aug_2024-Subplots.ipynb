```python
plt.subplot(1,2,1)
sns.countplot(x='continent',data = visa_df)
plt.subplot(1,2,2)
sns.countplot(x='case_status',data = visa_df)
```

```python
plt.subplot(2,1,1)
sns.countplot(x='continent',data = visa_df)
plt.subplot(2,2,2)
sns.countplot(x='case_status',data = visa_df)
```

```python
plt.figure(figsize=(14,3))
plt.subplot(1,2,1)
sns.countplot(x='continent',data = visa_df)
plt.subplot(1,2,2)
sns.countplot(x='case_status',data = visa_df)
```

```python
plt.figure(figsize=(14,10))
plt.subplot(2,2,1)
sns.countplot(x='continent',data = visa_df)
plt.subplot(2,2,2)
sns.countplot(x='requires_job_training',data = visa_df)
plt.subplot(2,2,3)
sns.countplot(x='has_job_experience',data = visa_df)
plt.subplot(2,2,4)
sns.countplot(x='case_status',data = visa_df)
```

```python
visa_df.dtypes
types = dict(visa_df.dtypes)
categorical_visa_df = [key for key,value in types.items() if value == 'object']
numerical_visa_df = [key for key,value in types.items() if value != 'object']
print(categorical_visa_df,'\n',numerical_visa_df)
```

```python
cols = categorical_visa_df[1:]
plt.figure(figsize=(16,14))
for i in range (len(cols)):
    plt.subplot(3,3,i+1)
    sns.countplot(x=cols[i],data=visa_df)
```
