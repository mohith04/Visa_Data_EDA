**Hint**
- whenever you want to create a list from a series
- we will need to convert series into dictionary
- when we convert we convert it will become a key value pair
- if that value == object type, then keep that in one list
- if it not= object in other list 

```python
visa_dtypes = visa_df.dtypes
visa_dtypes_dict = dict(visa_dtypes)
visa_dtypes_dict
```

```python
cat_list = []
num_list = []
for i,j in visa_dtypes_dict.items():
    if j == 'object':
        cat_list.append(i)
    else:
        num_list.append(i)

print (cat_list)
print (num_list)
```

```python
cat_list_comp = [i for i,j in visa_dtypes_dict.items() if j == 'object']
num_list_comp = [i for i,j in visa_dtypes_dict.items() if j != 'object']
print (cat_list_comp)
print (num_list_comp)
```
