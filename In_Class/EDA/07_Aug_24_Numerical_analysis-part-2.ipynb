**Imperical rule**
- (mean - 1std) and (mean + 1std) = 68%
- (mean - 2std) and (mean + 2std) = 95%
- (mean - 3std) and (mean + 3std) = 99.7%

- step 1: calculate the mean of prevailing wage 
- step 2: calculate the std of prevailing wage
- step 3: lb=mean-1*std
- step 4: ub=mean+1*std
- step 5: con1= data<ub
- step 6: con2 = data>lb
- step 7: con3 = con1 and con2
- step 8: data[con3]
- step 9: compare with 68% 25480

```python
mean_wage = visa_df['prevailing_wage'].mean()
sigma_wage = visa_df['prevailing_wage'].std()
low_b_sigma_1 = mean_wage - sigma_wage
upper_b_sigma_1 = mean_wage + sigma_wage
con_1_sigma_1 = visa_df['prevailing_wage']<upper_b_sigma_1
con_2_sigma_1 = visa_df['prevailing_wage']>low_b_sigma_1
con_sigma_1 = con_1_sigma_1 & con_2_sigma_1
len_of_68_per =len(visa_df[con_sigma_1])
# data_68_per = visa_df(con_sigma_1)
print (len_of_68_per)
print (68*visa_df/100)
```
