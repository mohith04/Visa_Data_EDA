```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

$SeaBorn$

- We can plot bar charts using many packages 
- In matplotlib we use **plt.bar**
- in seaborn we us **countplot**
- In matplotlib,
    - we will take raw data
    - we will create our frequency table i.e. dataframe
    - we will pass the column1, column2 and dataframe name
    - eg: Continent column ===> continent_df ===> plt.bar
- In Seaborn,
    - it takes raw data
    - we will provide direct column name
    - we will provide main dataframe name
    - eg: Continent column direclty ===> visa_df 

```python
visa_df = pd.read_csv('Visadataset.csv')
```

```python
plt.figure(figsize=(8,5))
order_continents = ['Asia','Europe','North America','South America','Africa','Oceania']
sns.countplot(data = visa_df,
             x = 'continent',
             order = order_continents)
plt.title('Count of Visa Applications by Continent')
plt.xlabel('Continent')
plt.ylabel('Count')
plt.savefig("No_of_application_vs_Continents.jpg")
plt.show()
```
