```python
for i in visa_df_unique:
    con = visa_df['continent'] == i
    count = len(visa_df[con])
    print (f'The number of applications from {i} is: {count}')
```

```python
visa_df_unique_continents = visa_df['continent'].unique()
count = []
for i in visa_df_unique_continents:
    con = visa_df['continent'] == i
    count.append (len (visa_df [con]))


cols = ['Continents','Number of applications']
continents_df = pd.DataFrame(zip(visa_df_unique,count),columns=cols)
```
