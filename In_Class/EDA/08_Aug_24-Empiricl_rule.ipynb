**Histogram**
- By applying the empirical rule we can observe slight deviation in the count of data
- Also mena is a slight deviation from the median
- this gives an indication the data does not follow Normal distribution
- to understand more clearly will visualize the data
- it is Numerical data, so the plot name is:**Histogram**
- A histogram is a graph between Class intervals and class interval frequency
- the plot is available in **matplotlib**
- the method name is **hist**

```python
plt.hist(visa_df['prevailing_wage'])
```

```python
wage_data = visa_df['prevailing_wage']
plt.hist(x=wage_data,bins=20)
```

- Histogram will return 3 values 
    1. interval count
        - if we have 20 intervals, so we have 20 interval counts
    2. bins
        - bis means the number of intervals
        - we have 20 intervals so we need 21 values
    3. It will display how many intervals are available
- for eg in the bins array the first two from one interval
    1. For the first interval
        - 2.1367 to 15962.543 the count is 3863
        - 3863 observations are available between 2.1367 to 15962.543
    2. For the second interval
        - 15962.543 to 31922.95 the count is 2175
        - 2175 observations are available between 15962.543 to 31922.95
    3. For the third interval
        - 31922.95 to 47883.3567 the count is 2676
        - 2676 observations are available between 31922.95 to 47883.3567
