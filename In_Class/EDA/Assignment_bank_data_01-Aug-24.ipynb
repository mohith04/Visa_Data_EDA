```python
bank_df_dtypes = dict(bank_df.dtypes)
bank_df_dtypes
```

```python
catego_bank = [i for i, j in bank_df_dtypes.items () if j == 'object']
numer_bank = [i for i, j in bank_df_dtypes.items() if j!='object']
print (catego_bank,'\n',numer_bank)
```

```python
bank_df_for_keys = [keys for keys,values in bank_df_dtypes.items ()]
bank_df_for_values = [values for keys,values in bank_df_dtypes.items ()]
```

```python
bank_df_keys = bank_df.select_dtypes(include='object').columns
bank_df_values = bank_df.select_dtypes(exclude='object').columns
```

```python
bank_df_dtypes_table = pd.DataFrame(zip(bank_df_for_keys,bank_df_for_values),
                                    columns=['Bank data names','Bank data dtypes'])
bank_df_dtypes_table.to_csv("bank_dtypes_table.csv",index=False)
bank_df_dtypes_table
```

```python
root_directory = os.getcwd()
new_folder = 'Bank_graph'
new_dir = os.path.join(root_directory, new_folder)
if not os.path.exists(new_dir):
 os.makedirs(new_dir)
 print(f"Directory '{new_dir}' created successfully.")
else:
 print(f"Directory '{new_dir}' already exists.")
```

```python
for i in bank_df_keys[1:]:
    plt.figure(figsize=(8,5))
    order_bank = bank_df_new[i].value_counts().keys()
    sns.countplot(data = bank_df_new,
                 x = i,
                 order = order_bank)
    plt.title('Bar chart')
    plt.savefig(f'{new_dir}\\{i}_seaborn.jpg')
    plt.show()
```
