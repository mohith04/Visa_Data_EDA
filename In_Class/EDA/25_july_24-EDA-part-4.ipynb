$take$
- first arguments: indices
- indices will take some values in array: in a list 
- it tries to extract the data 
- that values we provided represent columns or rows 
- Axis = 0 represents rows, by default the value will be zero if we don't specify it  
- Axis = 1 represents columns

```python
visa_df.take([10,23,48])
# this retrives 3 rows which we have given inside the list
```

```python
visa_df.take([1,2,8,10],axis=1)
# we need to have the amount to columns that we are giving in the list but if we give more than the no of columns it will give error
```

$loc-iloc$
- visa_df.iloc[rows,columns]
- visa_df.iloc[start:stop,start:stop]
- visa_df.iloc[start:stop:step,start:stop:step]
- visa_df.iloc[[list of rows],[list of columns]]

    - case-2: rows = [100:105] and columns = [2,5,7]
    - case-3: rows = [100:105] and columns = [2:7]
    - case-4: rows = [100,200,300] and columns = [2:7]
    - case-5: rows = [100,200,300] and columns = :
    - case-6: rows = : and columns = [2,5,7]
