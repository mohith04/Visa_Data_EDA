```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bank_df = pd.read_csv('bank.csv',sep = ';')
bank_df
```

```python
bank_df.dtypes
```

```python
bank_dict_dtypes = dict(bank_df.dtypes)
bank_dict_dtypes
```

```python
categorical_bank = [i for i,j in bank_dict_dtypes.items () if j == 'object']
numerical_bank = [i for i,j in bank_dict_dtypes.items() if j != 'object']
print (categorical_bank,'\n',numerical_bank)
```

```python
len(categorical_bank),len(numerical_bank)
```

```python
cat_bank_df = pd.DataFrame(categorical_bank,columns=['Catagorical items'])
cat_bank_df.to_csv("Catagorical_items_bank.csv",index=False)
cat_bank_df
```

```python
nume_bank_df = pd.DataFrame(numerical_bank,columns=['Numerical items'])
nume_bank_df.to_csv('Numerical_items_bank.csv',index=False)
nume_bank_df
```
